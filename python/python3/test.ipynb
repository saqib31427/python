```python
#multiplication of the matrix in n times 
def matrixpower(A,n):
 B=A
 C=A
 for i in range(n-1): 
  C[0][0]=A[0][0]*B[0][0]+A[0][1]*B[1][0]
  C[0][1]=A[0][0]*B[0][1]+A[0][1]*B[1][1]
  C[1][0]=A[1][0]*B[0][0]+A[1][1]*B[1][0]
  C[1][1]=A[1][0]*B[0][1]+A[1][1]*B[1][1]
  B=C
 return C
A = [[1,1],[1,0]]
print(matrixpower(A,2))



```

```python
#write a python program to count the number of words in a sentance and print the count of each word in the sentance
sentance = input("enter the sentance ")
words = sentance.split()
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
print("Number of words in the sentence:", len(words))
print("Count of each word:")
for word, count in word_count.items():
    print(f"{word}: {count}")
#now we have count of each word in the sentance and we can use this count to find the most common word in the sentance or we can use this count to find the least common word in the sentance or we can use this count to find the average count of words in the sentance etc.
    
```

```python
#write a python program to count the number of alphabetic characters in a sentance and print the count of each word in the sentance
#isalpha method is used to check if the character is an alphabetic character or not and it returns True if the character is an alphabetic character and it returns False if the character is not an alphabetic character and it is used to count the number of alphabetic characters in a sentance and print the count of each word in the sentance
sentance = input("enter the sentance ")
char_count = {}
for char in sentance:
    if char.isalpha():
        char_count[char] = char_count.get(char, 0) + 1
print("Count of each alphabetic character:")
for char, count in char_count.items():
    print(f"{char}: {count}")
```
